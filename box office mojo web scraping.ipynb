```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
movies_table = driver.find_element(By.XPATH, "//div[@class='a-section imdb-scroll-table-inner']//tbody")
```

```python
movies = movies_table.find_elements(By.XPATH, ".//tr")
```

```python
data = []
for movie in movies:
    try:
        rank = movie.find_element(By.XPATH, ".//td[@class='a-text-right mojo-header-column mojo-truncate mojo-field-type-rank']").text
        title = movie.find_element(By.XPATH, ".//td[@class='a-text-left mojo-field-type-title']").text
        lifetime_gross = movie.find_element(By.XPATH, ".//td[@class='a-text-right mojo-field-type-money']").text
        release_year = movie.find_element(By.XPATH, ".//td[@class='a-text-left mojo-field-type-year']").text
        
        data.append({
            'Rank': rank,
            'Title': title,
            'Lifetime Gross': lifetime_gross,
            'Release Year': release_year
        })

        print(f"Rank: {rank}")
        print(f"Title: {title}")
        print(f"Lifetime Gross: {lifetime_gross}")
        print(f"Release Year: {release_year}")
        print("-" * 40)

    except Exception as e:
        print("Error extracting data:", e)
        continue
df = pd.DataFrame(data)
csv_file = 'box_office_mojo.csv'
df.to_csv(csv_file, index=False, encoding='utf-8')
```
